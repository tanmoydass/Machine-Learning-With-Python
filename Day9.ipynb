```python
#LINEAR REGRESSION - MODEL 2(We CREATE OUR OWN DATASET)
import numpy as np
experience = np.arange(1,11)#It will print numbers from 1 to 10
np.random.seed(35)
#random.seed keep my values constant
salary = np.random.randint(20000,100000,10)
salary = np.sort(salary)
print(experience)
print(salary)
```

```python
#3.Visualisation
import matplotlib.pyplot as plt
plt.scatter(experience,salary)
```

```python
#4.Divide into input and output
#Input has to be always 2 dimensional
x = experience.reshape(-1,1)
x
```

```python
x.ndim
```

```python
y=salary
y
```

```python
#we are not performing no 5 and no 6
```

```python
#RUN a CLASSIFIER/REGRESSOR/CLUSTERER
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
#Predict the output
y_pred=model.predict(x) #using the input values , we predict the output
y_pred #Predicted values
```

```python
#Before Normalisation

import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x1=np.random.randint(1,50,30)
x1=np.sort(x1)
x2=np.random.randint(10000,70000,30)
plt.plot(x1,x2)
```

```python
#These graphs are no where related to machine learning
#Here we are only understanding the concept of scalling, using the graphs

#2.After Normalisation
x1min= min(x1)
x1max= max(x1)
x2min= max(x2)
x2max= max(x2)
x1norm= (x1-x1min)/(x1max-x1min)
x2norm= (x2-x2min)/(x2max-x1min)
plt.plot(x1norm,x2norm)
```

```python
#We want to consider only the numeric data
#So we will create a new dataframe with only numeric data
df_numeric=df.select_dtypes(include=["int64","float64"])
df_numeric
```

```python
#Now we have to remove or drop the CarID and Symbling column
df_numeric=df_numeric.drop(["symboling","car_ID"],axis=1) #axis = 1 -column,axis = 0 - row
df_numeric
```

```python
#divide the data into i/p and o/p
#output - price
#input - All the columns except the price column
```

```python
x=df_numeric.iloc[:,0:13].values
x
```

```python
y=df_numeric.iloc[:,13].values
y
```

```python
#5.TRAIN and TEST VARIABLES
#sklearn.model_selection - package , train_test_split - library
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0)
#Whatever data splitting /data allocation happens to the xtrain,x_test,ytrain,ytest variables , we want those allocated values to remain constant
#By default the training variables get 75 % and testing variables get 25%
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#6. SCALING or NORMALISATION -DONE ONLY FOR INPUTS
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)
```

```python
#7.RUN a CLASSIFIER/REGRESSOR/CLUSTERER
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#8.MODEL FITTING
model.fit(x_train,y_train)
```

```python
#9.PREDICT THE OUTPUT
y_pred=model.predict(x_test) #By taking the input testing data , we predict the output
y_pred
```
